# Axes, Lables, Grid aspects of a plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[23,25,27,28,29,31,34]
min_t=[25,30,32,35,38,40,42]
avg_t=[35,40,42,44,48,50,52]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather data")

plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.grid()

plt.legend()
```

# Histograms

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]

plt.hist(blood_sugar,bins=[80,100,120,140],rwidth=0.99,color='green')
```

## multi bar histogram

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[100,65,98,180,139,98,102,105,115,85,67,72,139]


plt.title('Sugar levels by men and women')
plt.xlabel('level')
plt.ylabel('values')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,120,140],rwidth=0.99,color=['green','violet'], label=['men','women'])
plt.legend()
```

## horizontal histogram

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[100,65,98,180,139,98,102,105,115,85,67,72,139]


plt.title('Sugar levels by men and women')
plt.xlabel('Level')
plt.ylabel('values')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,120,140],rwidth=0.99,
         color=['green','violet'], label=['men','women'], orientation='horizontal')
plt.legend()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values = [510,10,25,20]
exp_label = ['rent','wifi','utilities','other']
```

```python
plt.title("Expenses")
plt.pie(exp_values,labels=exp_label,radius=0.7, autopct='%0.2f%%')
plt.legend()
plt.show()
```
